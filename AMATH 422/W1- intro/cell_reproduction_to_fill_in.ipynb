First, import packages, and give their contents a prefix name (that's the "as").  We'll always do this. 

```python
import matplotlib.pylab as plt   # That gives plotting, and the next line makes plots appear inline in notebook
%matplotlib inline  
import numpy as np  # That gives numerical arrays 
```

Define parameters

```python
p= 1/2 ###TO FILL IN ###; #fraction of cells that survive each reproduction
num_generations= 50 ###TO FILL IN ###;  #a total of that many generations, including generation 0
```

Define an numpy array.  
These are lists of numbers that we will operate on.  
We'll define an array of zeros, with generation_end+1 entries

```python
n_arr=np.zeros(num_generations)
print(n_arr) #use print to display a "variable"
```

Initialize the first entry in n_arr to a number of individuals in generation 0.  
Note:  Python uses [] to index into arrays, and INDEXING STARTS AT 0.

```python
# Just saying that we are starting with 50 cells
n_arr[0]= 50###TO FILL IN ###
```

Now we'll make a for loop, to run our dynamical model!  
First, we need to define what we'll be looping over.  
This is the iterator t_arr, which counts which generation we are on.  t stands for time here.  
np.arange(start,stop), where start and stop are two integers, makes list from start to stop-1

```python
t_arr=np.arange(0,num_generations-1)
print(t_arr)
```

## Run the dynamical model!

```python
# We are saying that we are quadrupling the number of cells 
# Number of cells at the next time step = 4 * p * (# of cells in the previous time step)
for t in t_arr:
        n_arr[t+1]= 4 * p * n_arr[t]###TO FILL IN ###
```

Plot!

```python
plt.plot(n_arr)
plt.xlabel('generation')
plt.ylabel('number of cells')
plt.title('cell reproduction')
```

### Now move to the nonlinear cell reproduction model

```python
p=.9; #fraction of cells that survive each reproduction
num_generations=200;  #a total of that many generations
K=2000;  #reproduction capacity

n_arr=np.zeros(num_generations)
n_arr[0]=1000

t_arr=np.arange(0,num_generations-1)
```

## Run the dynamical model!

```python
for t in t_arr:
        n_arr[t+1]=4*n_arr[t]*p*(1-n_arr[t]/K) 
```

```python
plt.plot(n_arr)
plt.xlabel('generation')
plt.ylabel('number of cells')
plt.title('nonlinear cell reproduction')
```
