```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python

# Define the ODE function
def dydt(t, y):
    
    return t/3 + np.sin(y)
    # return -2 * y + np.exp(-t)

# Initial condition
y0 = -2

# Time interval
t_span = (0, 20)

# Solve the ODE
sol = solve_ivp(dydt, t_span, [y0])

# Plot the solution
plt.plot(sol.t, sol.y[0])
plt.xlabel('t')
plt.ylabel('y')
plt.title('Solution to the ODE')
plt.grid(True)

plt.show()
```

```python
# def ode_fun(t, y):
#     dy = np.zeros(3)
#     dy[0] = y[1] * y[2]  # y1' = y2 * y3
#     dy[1] = -y[0] * y[2]  # y2' = -y1 * y3
#     dy[2] = -0.51 * y[0] * y[1]  # y3' = -0.51 * y1 * y2
#     return dy



# Define the system of ODEs
def ode_fun(t, z):
    y, v = z
    dydt = v
    dvdt = y**2 - y
    return [dydt, dvdt]

# Initial conditions
y0 = 0.5
v0 = 1.0
z0 = [y0, v0]

# Time span
t_span = [0, 3]
t_eval = np.linspace(0, 3, 300)  # For smooth plots

# Solve the system of ODEs
sol = solve_ivp(ode_fun, t_span, z0, t_eval=t_eval)

# Extract solutions
t = sol.t
y = sol.y[0]
v = sol.y[1]
y_hat = "ŷ" 

# Plot y vs. t
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(t, y)
# plt.plot(t, )
plt.title('y vs. t')
plt.xlabel('t')
plt.ylabel('y')
plt.grid(True)
# plt.legend()

# Plot v (y') vs. y (Phase plane)
plt.subplot(1, 2, 2)
plt.plot(y, v)
plt.title(f'Phase Plane {y_hat} vs. y)')
plt.xlabel('y')
plt.ylabel(f"{y_hat}")
plt.grid(True)
# plt.legend()

plt.tight_layout()
plt.show()
```
