```python
import numpy as np 
import matplotlib.pyplot as plt
import scipy.io
import pandas as pd
from scipy.integrate import solve_ivp
```

```python
# Define the ODE function
def dvdt(t, v):
    g = 9.8
    c = 0.3773

    return (g - c * v**2 )

# Initial condition
y0 = 0

# Time interval
t_span = (0, 5)

# Solve the ODE
sol = solve_ivp(dvdt, t_span, [y0])

# Plot the solution
plt.plot(sol.t, sol.y[0])
plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")
plt.title("Solution of dv/dt = g - c*v^2 with v(0) = 0")
plt.grid(True)

plt.show()
```
