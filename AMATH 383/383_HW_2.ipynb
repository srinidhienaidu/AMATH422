```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Orbital period (T) in Earth years and semimajor axis (a) in AU for planets and dwarf planets
orbital_periods = np.array([0.241, 0.615, 1.0, 1.881, 11.86, 29.46, 84.01, 164.8, 248, 4.60, 283.8, 305.3, 558])  # T (Earth years)
semimajor_axes = np.array([0.387, 0.723, 1.0, 1.524, 5.203, 9.537, 19.19, 30.07, 39.48, 2.77, 43.34, 45.79, 67.74])  # a (AU)

# Logarithms of the data
log_T = np.log(orbital_periods)
log_a = np.log(semimajor_axes)
# print('log_T', log_T)
# print('log_a', log_a)

# Linear regression (log(T) vs log(a)) to find the slope (which is the power b)
slope, intercept, r_value, p_value, std_err = linregress(log_a, log_T)
labels = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune', 'Pluto', 'Ceres', 'Haumea', 'Makemake', 'Eris']

# Generate the log-log plot
plt.figure(figsize=(8, 6))
plt.scatter(log_a, log_T, color='blue', label='Data points')
for i, label in enumerate(labels):
    plt.annotate(label, (log_a[i], log_T[i]), textcoords="offset points", xytext=(1,-1), ha='right')

# Plot the regression line
plt.plot(log_a, slope * log_a + intercept, color='red', label=f'Linear fit (slope = {slope:.2f})')

# Annotating the plot
plt.title('Log-Log Plot of Orbital Period vs Semimajor Axis')
plt.xlabel('(Semimajor Axis (AU))')
plt.ylabel('(Orbital Period (Years))')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()

# Output the slope and statistical data
slope, intercept, r_value, std_err
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
a = 1  # Example value for a
M = 1  # Example value for M (adult mass)
K = 1  # Initial condition for R(t)

# Time array (t) and nondimensionalized time (tau)
t = np.linspace(0, 10, 100)  # Time from 0 to 10
tau = (a * t) / (4 * M**(1/4))  # Non-dimensionalized time

# Calculate R(t)
R = K * np.exp(-tau)

# Calculate ln(R(t))
ln_R = np.log(R)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(tau, ln_R, label='ln(R(t))', color='blue')
plt.title('Plot of ln(R(t)) vs. Non-Dimensionalized Time ')
plt.xlabel('Non-Dimensionalized Time  = at / (4M^(1/4))')
plt.ylabel('ln(R(t))')

plt
```
