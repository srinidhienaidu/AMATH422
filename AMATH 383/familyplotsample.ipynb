```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def family_plot_sample():
    q_0 = np.arange(1, 10)  # Parameter we want to vary: c = 1, 2, ..., 20.
    t = np.arange(-8, 8.01, 0.01)  # Range of the independent variable to plot: t = -8, -7.99, -7.98, ..., 8.

    plt.figure(1)  # First figure
    for p in range(len(q_0)):  # For every value of c
        y = q_0[p] * np.exp(-t)
        # print(y)
        # y = np.exp(-t**2 / c[p])  # Dependent variable
        plt.plot(t, y, color = '#1f77b4')  # Plot with red line

    plt.xlabel('t')
    plt.ylabel('q(t)')
    plt.title('q_0 * e^{-t} for c=1:1:20')
    plt.show()

family_plot_sample()
```

```python
# Define parameters
R = 1
C = 1
t = np.linspace(0, 10, 400)  # Time from 0 to 10 seconds

# Define a function for q(t)
def q(t, q0, v, R=1, C=1):
    return (q0 - v * C) * np.exp(-t / (R * C)) + v * C

# Initial condition and range of v values
q0 = 1
v_values = np.linspace(0, 5, 5)  # Different values of v

# Plot the family of solutions
plt.figure(figsize=(8, 6))
for v in v_values:
    plt.plot(t, q(t, q0, v), label=f'v={v}')

plt.title('Charge decay for different values of v with R = 1, C = 1')
plt.xlabel('Time (t)')
plt.ylabel('Charge (q(t))')
plt.legend()
plt.grid(True)
plt.show()
```
